```python
cat_cols = [
    'user_id',
    'REGION',
    'TENURE',
    'MRG',
    'TOP_PACK',
]

numeric_cols = [
    'MONTANT',
    'FREQUENCE_RECH',
    'REVENUE',
    'ARPU_SEGMENT',
    'FREQUENCE',
    'DATA_VOLUME',
    'ON_NET', 
    'ORANGE',
    'TIGO',
    'ZONE1',
    'ZONE2',
    'REGULARITY',
    'FREQ_TOP_PACK',
]

target = 'CHURN'
```

```python
train[target].value_counts()
```

```python
# region
for reg in train['REGION'].unique():
    print('=' * 5, reg, '=' * 5)
    try:
        t = train[train['REGION'] == reg][target].value_counts()
        print('0:', t[0])
        print('1:', t[1])
        print('%', f'{t[1]/t[0] * 100:2f}%')
    except Exception as e:
        print(e)
```

```python
train['REGION'].value_counts()
```

```python
# tenure -- ordinal feature
for tenure in train['TENURE'].unique():
    print('=' * 5, tenure, '=' * 5)
    try:
        t = train[train['TENURE'] == tenure][target].value_counts()
        print('0:', t[0])
        print('1:', t[1])
        print('%', f'{t[1]/t[0] * 100:2f}%')
    except Exception as e:
        print(e)
```

```python
mapping = {
    'D 3-6 month': 0,
    'E 6-9 month': 1,
    'F 9-12 month': 2,
    'G 12-15 month': 3,
    'H 15-18 month': 4,
    'I 18-21 month': 5,
    'J 21-24 month': 6,
    'K > 24 month': 7,
}
train['TENURE_ORD'] = train['TENURE'].map(mapping)
train[['TENURE_ORD']].corrwith(train[target])
```
